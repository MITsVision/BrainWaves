```python
import pandas as pd
import numpy as np
import scipy as sc
import sklearn as skl
import matplotlib
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
df = pd.read_csv("./train.csv")
```

```python
df['zero'] = np.ones(9366) # for aditional theta(0) to make Q0x0 + Q1x1 + ...
x = df[['zero','sold','euribor_rate','libor_rate','bought']]
y = df['return']
m = len(x)
alpha = 0.001
rep = 1000
theta = [0.5,0.5,0.5,0.5,0.5]
```

```python
for me in range(rep):
    mulm = np.dot(x,theta)
    diff_h_y = np.subtract(mulm,y)
    for i in range(5):
        all_sum = np.sum(diff_h_y * x[x.columns[i]])
        theta[i] = theta[i] - (alpha * np.divide(all_sum,m))
    print("iteration:{}".format(me))
    print(theta)
```

```python
mulm = np.dot(x,theta)
diff_h_y = np.subtract(mulm,y)
for i in range(5):
    all_sum = np.sum(diff_h_y * x[x.columns[i]])
    print(all_sum)
```
